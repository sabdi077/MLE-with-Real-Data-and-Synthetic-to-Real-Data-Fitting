```python
from NewOOP2 import Runs 
import numpy as np
import matplotlib.pyplot as plt

'''Generating random/real alphas and betas'''
def generate_random_alpha_beta(s):
    alphas = np.random.uniform(0.2, 1, s)
    betas = np.random.uniform(0.2, 9, s)
    return alphas, betas
'''
Num_mice:=The number of randomly generated alpha and beta
num_est:=The number of estimates for each randomly generated alpha and beta
Num_trials:=The number of states, actions, and rewards for each randomly generated alpha and beta
'''
Num_mice= 50   
num_est = 50
Num_trials=1000

# Generate N random alpha and beta values
real_alphas, real_betas = generate_random_alpha_beta(Num_mice)

# Create an instance of RunsOverSimulation
runs_simulation = Runs(Num_trials, Num_mice, num_est)

# Estimate alphas and betas for each randomly generated alpha and beta
alpha_estimates_list, beta_estimates_list = runs_simulation.estimate_alphas_betas(real_alphas, real_betas)

'''Mean PLOTS'''
plt.figure(figsize=(8, 6))
for j in range(num_est):
    plt.plot(real_alphas[j], np.mean(alpha_estimates_list[j]), 'bo', alpha=0.5)

plt.xlabel('Randomly Generated Alphas')
plt.ylabel('Alpha Estimates')
plt.title('Alpha Estimates for Randomly Generated Alphas')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
for j in range(num_est):
    plt.plot(real_betas[j], np.mean(beta_estimates_list[j]), 'ro', alpha=0.5)

plt.xlabel('Randomly Generated Betas')
plt.ylabel('Beta Estimates')
plt.title('Beta Estimates for Randomly Generated Betas')
plt.grid(True)
plt.show()

'''Variance PLOTS'''
plt.figure(figsize=(8, 6))
for j in range(num_est):
    plt.plot(real_alphas[j], np.var(alpha_estimates_list[j]), 'bo', alpha=0.5)

plt.xlabel('Randomly Generated Alphas')
plt.ylabel('Alpha Estimates Variance')
plt.title('Alpha Estimates Variance for Randomly Generated Alphas')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
for j in range(num_est):
    plt.plot(real_betas[j], np.var(beta_estimates_list[j]), 'ro', alpha=0.5)

plt.xlabel('Randomly Generated Betas')
plt.ylabel('Beta Estimates Variance')
plt.title('Beta Estimates Variance for Randomly Generated Betas')
plt.grid(True)
plt.show()


''' MSE PLOTS'''
plt.figure(figsize=(8, 6))
MSE_alpha = []

for j in range(num_est):
    mean_estimate = np.mean(alpha_estimates_list[j])
    squared_errors = np.square(alpha_estimates_list[j] - mean_estimate)
    mse_value = np.mean(squared_errors)
    MSE_alpha.append(mse_value)
    plt.plot(real_alphas[j], MSE_alpha[j], 'bo', alpha=0.5)

plt.xlabel('Randomly Generated Alphas')
plt.ylabel('Alpha Estimates Variance (MSE)')
plt.title('Alpha Estimates Variance (MSE) for Randomly Generated Alphas')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
MSE_beta = []

for j in range(num_est):
    mean_estimate_beta = np.mean(beta_estimates_list[j])
    squared_errors_beta = np.square(beta_estimates_list[j] - mean_estimate_beta)
    mse_value_beta = np.mean(squared_errors_beta)
    MSE_beta.append(mse_value_beta)
    plt.plot(real_betas[j], MSE_beta[j], 'ro', alpha=0.5)

plt.xlabel('Randomly Generated Betas')
plt.ylabel('Beta Estimates Variance (MSE)')
plt.title('Beta Estimates Variance (MSE) for Randomly Generated Betas')
plt.grid(True)
plt.show()
```
